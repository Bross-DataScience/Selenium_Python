```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
wd = webdriver.Chrome()
```

```python
## site alvo -- Google
## cotação dólar
wd.get('https://www.google.com/')

## acessando google e inputando valor na campo de pesquisa através do Xpath
wd.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dolar")

## confirmando pesquisa, pressionando botão Enter através do Xpath
wd.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

## pegando valor do dólar através do Xpath
rate_dolar = wd.find_element_by_xpath(
    '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(f'Cotação Dolar Hoje: {rate_dolar}')

## site alvo -- Google
# cotação Euro
wd.get('https://www.google.com/')

## acessando google e inputando valor na campo de pesquisa através do Xpath
wd.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro")

## confirmando pesquisa, pressionando botão Enter através do Xpath
wd.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

## pegando valor do dólar através do Xpath
rate_euro = wd.find_element_by_xpath(
    '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(f'Cotação Euro Hoje: {rate_euro}')

## site alvo -- MelhorCambio
# cotação Ouro
wd.get('https://www.melhorcambio.com/ouro-hoje')

## pegando valor atráves do Xpath
rate_ouro = wd.find_element_by_xpath(
    '//*[@id="comercial"]').get_attribute('value')

## limpando formato de "," para "."
rate_ouro = rate_ouro.replace(',', '.')
print(f'Cotação Ouro Hoje: {rate_ouro}')
```

```python
import pandas as pd

df = pd.read_excel('produtos.xlsx')
df.head(5)
```

```python
## atualizar cotação
df.loc[df['Moeda'] == 'Dólar', 'Cotação'] = float(rate_dolar)
df.loc[df['Moeda'] == 'Euro', 'Cotação'] = float(rate_euro)
df.loc[df['Moeda'] == 'Ouro', 'Cotação'] = float(rate_ouro)

## atualizar o preço de compra -- preço original * cotação
df['Preço de Compra'] = df['Preço Base Original'] * df['Cotação']

## atualizar o preço de venda -- preço de compra * margem
df['Preço de Venda'] = df['Preço de Compra'] * df['Margem']

## formatando para duas casas decimais as seguintes colunas
df['Cotação'] = df['Cotação'].map('{:.2f}'.format)
df['Preço Final'] = df['Preço Final'].map('{:.2f}'.format)
df['Preço de Venda'] = df['Preço de Venda'].map('{:.2f}'.format)
df['Preço de Compra'] = df['Preço de Compra'].map('{:.2f}'.format)

df.head(5)
```

```python
df.to_excel('new_produtos.xlsx', index=False)
wd.quit()
```
